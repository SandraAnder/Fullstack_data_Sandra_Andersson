## f) Graph different data engineers salary. There are many approaches to graph this, try different and reason about pros and cons of each you choose.

```python
engineers_salary = df.loc[
    df['Position'].isin(['Junior Data Engineer', 'Senior Data Engineer', 'Big Data Engineer']), ['Position', 'Salary_SEK']]

engineers_salary = engineers_salary.sort_values(by='Salary_SEK', ascending=True)
engineers_salary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for position in engineers_salary['Position'].unique():
    subset = engineers_salary[engineers_salary['Position'] == position]
    plt.plot(subset['Salary_SEK'].values, marker='o', label=position)

plt.title('Salaries of different Data Engineers at the same company')
plt.ylabel('Salary in SEK')
plt.legend(title='Position')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,4))

positions = ["Junior Data Engineer", "Senior Data Engineer", "Big Data Engineer"]
colors = ['plum', 'magenta', 'mediumorchid']

for i, position in enumerate(positions):
    subset = engineers_salary[engineers_salary['Position'] == position]
    axes[i].bar(range(len(subset)), subset['Salary_SEK'], color=colors[i])
    axes[i].set_title(f'Salary for {position}')
    axes[i].set_ylabel('Salary in SEK')

plt.tight_layout()
```
